```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
```

```python
logreg = LogisticRegression(C=1e5)
logreg.fit(X, y)
```

```python
# Generate Data test from Min & Max
# Coi toàn bộ các điểm từ Min - Max (cách nhau 0.05) là Data test
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = .02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])# Xác định đường boundary giữa 3 classes

Z = Z.reshape(xx.shape)# Reshape Z from 1D to 2D
plt.contourf(xx, yy, Z, cmap='Paired')

plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='Paired')

plt.xlabel('Length')# Set name of axis
plt.ylabel('Width')
plt.xticks(())# Hide ticks of axis
plt.yticks(())
plt.show()
```
