```python
import numpy as np
from numpy import linalg as LA

m, n = 2, 3
A = np.random.rand(m, n)

U, S, V = LA.svd(A)

# checking if U, V are orthogonal and S is a diagonal matrix with
# non-negative decreasing elements
print('Frobenius norm of (UU^T - I) =', LA.norm(U.dot(U.T) - np.eye(m)))
print('S =', S)
print('Frobenius norm of (VV^T - I) =', LA.norm(V.dot(V.T) - np.eye(n)))
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

img = mpimg.imread('building2.jpg')
plt.imshow(img)
imgplot = plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# to gray
gray = 0.2125*img[:, :, 0] + 0.7154*img[:, :, 1] + 0.0721*img[:, :, 2]
plt.imshow(gray)
plt.axis('off')
plt.show()
```

```python
from numpy import linalg as LA
U, S, V = LA.svd(gray)

plt.semilogy(S)
plt.xlabel('$k$', fontsize=20)
plt.ylabel('$\sigma_k$', fontsize=20)
plt.tick_params(axis='both', labelsize=18)
plt.tight_layout()
plt.show()
```

```python
a = np.sum(S**2)
b = np.zeros_like(S)
for i in range(S.shape[0]):
    b[i] = np.sum(S[:i+1]**2, axis=0)/a
    
plt.plot(b)
plt.xlabel('$k$', fontsize=20)
plt.ylabel('$\|\mathbf{A} - \mathbf{A}_k\|_F^2 / \|\mathbf{A}\|_F^2$', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=18)
plt.tight_layout()
plt.show()
```

```python
e = 1-b

def approx_rank_k(U, S, V, k):
    Uk = U[:, :k]
    Sk = S[:k]
    Vk = V[:k, :]
    return np.around(Uk.dot(np.diag(Sk)).dot(Vk))
A1 = []
for k in range(5, 100, 5):
    A1.append(approx_rank_k(U, S, V, k))
```

```python
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
def update(i):
    ani = plt.cla()
    ani = plt.imshow(A1[i])
    label = '$k$ = %d: error = %.4f' %(5 * i + 5, e[i])
    ax.set_xlabel(label)
    ani = plt.axis('off')
    ani = plt.title(label)
    return ani, ax
anim = FuncAnimation(fig, update, frames=np.arange(0, len(A1)), interval=500)

from IPython.display import HTML
HTML(anim.to_jshtml())
```
